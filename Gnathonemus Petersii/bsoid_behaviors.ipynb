```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import os 
import glob

os.chdir(r'D:\datos_GPetersii\datos_GPetersii\Fish1\Object\raw\50fps\B-soid\BSOID')
files = glob.glob('*.csv')
```

```python
track = pd.read_csv(files[0]).iloc[:, 1:]
track.head()
```

```python
cols = ['B-soid']
for col in track.columns[1:]:
    col = col[1:-1]
    cols.append(tuple(col.split(', ')))

track.columns = cols
track.head()
```

```python
xpos = pd.DataFrame()
ypos = pd.DataFrame()

for col in track.columns[1:]:
    if col[2] == "'x'":
        xpos[col[1][1:-1]] = track[col]
    elif col[2] == "'y'":
        ypos[col[1][1:-1]] = track[col]
    
```

```python
track.head()
```

```python
min_frames = 50
behav = 4
frames =[]
start = 0
end = 0

for i,row in track.iterrows():
    if row[0] == behav:
        if start == None:
            start = i
        end = i 
    else:
        if start is not None and (end - start + 1) >= min_frames:
            frames.append((start, end))
        start = None

frames
```

```python
num = frames[0][1] - frames[0][0]
cmap = sns.color_palette("ch:s=0.9,r=-0.55", as_cmap=True)
colors = [cmap(i/num) for i in range(num)]

fig, ax = plt.subplots()
for i in range(frames[0][0], frames[0][1]):
    plt.scatter(xpos.iloc[i,:]/12, ypos.iloc[i,:]/12, c=colors[i-frames[0][0]])
    plt.plot(xpos.iloc[i,:]/12, ypos.iloc[i,:]/12, c=colors[i-frames[0][0]], linestyle = '--')

#fig.savefig()
```
